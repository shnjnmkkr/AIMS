```python
#training data used =https://www.kaggle.com/datasets/spscientist/students-performance-in-exams
import pandas as pd
path=input("Enter the path of the dataset")
df=pd.read_csv(path)
df.head()
```

```python
def OrdinalEncoder(df):
    df_new=df.copy()
    mapping={}
    s = (df.dtypes == 'object')
    object_cols = list(s[s].index)
    if len(object_cols) == 0:
        print("No categorical columns found.")
        return df
    print("The Columns containing Categorical variables are:",object_cols)

    for col in object_cols:
        print(f"The column {col} has {df[col].unique()} as unique values")
        
    while True:
        print("\nSelect a column to be encoded ordinally, or type 'exit' to finish: ")
        catCol = input()
        if catCol == 'exit':
            break
        if catCol not in object_cols:
            print(f"{catCol} is not a valid categorical column. Please select a valid column.")
            continue
        mapping[catCol] = {}
        print(f"Encoding values for column '{catCol}'")
        for cat in df[catCol].unique():
            while True:
                try:
                    encoded_value = int(input(f"Enter the encoded value for {cat} between 0 (bottom of the hierarchy) to {len(df[catCol].unique())-1} (top of the hierarchy): "))
                    if 0 <= encoded_value <= len(df[catCol].unique()) - 1:
                        mapping[catCol][cat] = encoded_value
                        break
                    else:
                        print(f"Please enter a value between 0 and {len(df[catCol].unique())-1}.")
                except ValueError:
                    print("Please enter a valid integer.")
        df_new[catCol] = df[catCol].replace(mapping[catCol])
        print(df_new.head())
    return df_new
    #~shnjnmkkr
```
